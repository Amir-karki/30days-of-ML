```python
# importing the libraries
import pandas as pd
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
```

```python
# reading the csv file from the local directory
url = r'C:\Users\12368\OneDrive\Desktop\DataScience\melb_data.csv'
data = pd.read_csv(url)
```

```python
# general description of the data
data.describe()
```

```python
# check the infos of the data, check the # of rows, # of cols, total entries, DTypes and so on.
data.info()
```

```python
# checks if the data contains any null values
null_Check = data.isnull().any()
```

```python
# selects the columns that contains these null values 
cols_with_nulls = [cols for cols in data.columns if data[cols].isnull().any()]
cols_with_nulls
```

```python
# drops all the columns with null values
reduced_cols = data.dropna()
reduced_cols
```

```python
reduced_cols.info()
```

```python
# only selects the numerical columns, it will be used as a features to train the model
numerical_cols = reduced_cols.select_dtypes(include=['int64', 'float64'])
numerical_cols
```

```python
y = numerical_cols['Price']  # selects the target variable
X = numerical_cols.drop('Price', axis=1) # selects the predictors
print(f"Target variable y and predictors X")
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) #splits the dataset into training and testing set. 80% training set and 20% testing set
```

```python
# initialize the model and make predictions
model = DecisionTreeRegressor()
model.fit(X_train, y_train)
pred = model.predict(X_test)
mae = mean_absolute_error(y_test, pred)
print(f"The prediction is: {pred} with mean absolute error of {mae}.")
```
